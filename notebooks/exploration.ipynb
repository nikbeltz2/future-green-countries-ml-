```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

yr_of_data = pd.read_csv("Best Copy of Energy Data for CS ML 2000-2022 - Sheet1.csv")
yr_of_data.head()
```

```python
y = yr_of_data["co2_emissions_total"]
y
```

```python
X = yr_of_data.drop(columns=["country", "co2_emissions_total", "gdp_per_capita", "co2_emissions_per_capita", "elec_gen_hydro", "elec_gen_tide", "elec_gen_geothermal", "elec_gen_solar_thermal"])
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor

RF_model = RandomForestRegressor(n_estimators=200, random_state=40)
RF_model.fit(X_train, y_train)

RandomForestRegressor

```

```python
y_rf_train_pred = RF_model.predict(X_train)
y_rf_test_pred = RF_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(["Random Forest Regressor", rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ["Method", "Training MSE", "Training R2", "Testing MSE", "Testing R2"]
rf_results
```

```python
import pandas as pd

importance = RF_model.feature_importances_

feature_importance = pd.Series(importance, index=X.columns)
print(feature_importance.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_rf_test_pred)
plt.xlabel("Actual CO2")
plt.ylabel("Predicted CO2")
plt.show()
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

data2 = pd.read_csv("Best Copy of Energy Data for CS ML 2000-2022 - Sheet1.csv")
data2.head()
```

```python
data2["country_code"] = data2["country"].astype("category").cat.codes
```

```python
data2["renew/sus_elec_gen"] = data2["elec_gen_solar"] + data2["elec_gen_wind"] + data2["elec_gen_hydro/tide"] + data2["elec_gen_sol/geothermal"]
```

```python
data2["elec_gen_total"] = data2["renew/sus_elec_gen"] + data2["elec_gen_oil"] + data2["elec_gen_coal"] + data2["elec_gen_biofuels"] + data2["elec_gen_natural_gas"] + data2["elec_gen_nuclear"] + data2["elec_gen_other"] + data2["elec_gen_waste"]
```

```python
data2["true_renew_share"] = data2["renew/sus_elec_gen"] / data2["elec_gen_total"]
```

```python
data2
```

```python
y2 = data2["true_renew_share"]
y2
```

```python
X2 = data2.drop(columns=["country", "co2_emissions_total", "elec_gen_hydro", "elec_gen_tide", "elec_gen_geothermal",
                              "elec_gen_solar_thermal", "elec_gen_sol/geothermal", "elec_gen_hydro/tide", "elec_gen_solar", "elec_gen_wind", "true_renew_share", "elec_gen_total", "renew/sus_elec_gen",
                              "co2_emissions_total", "co2_emissions_coal", "co2_emissions_oil", "co2_emissions_natural_gas", "co2_emissions_other", "co2_emissions_elec_gen_incl_chp", 
                              "elec_gen_coal", "elec_gen_oil", "elec_gen_natural_gas", "elec_gen_biofuels", "elec_gen_waste", "elec_gen_nuclear", "elec_gen_other", "country_code"])
X2
```

```python
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

RF_model2 = RandomForestRegressor(n_estimators=200, random_state=42)
RF_model2.fit(X2_train, y2_train)
```

```python
y2_rf_train_pred = RF_model2.predict(X2_train)
y2_rf_test_pred = RF_model2.predict(X2_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rf2_train_mse = mean_squared_error(y2_train, y2_rf_train_pred)
rf2_train_r2 = r2_score(y2_train, y2_rf_train_pred)

rf2_test_mse = mean_squared_error(y2_test, y2_rf_test_pred)
rf2_test_r2 = r2_score(y2_test, y2_rf_test_pred)
```

```python
rf2_results = pd.DataFrame(["Random Forest Regressor", rf2_train_mse, rf2_train_r2, rf2_test_mse, rf2_test_r2]).transpose()
rf2_results.columns = ["Method", "Training MSE", "Training R2", "Testing MSE", "Testing R2"]
rf2_results
```

```python
import pandas as pd

importance2 = RF_model2.feature_importances_

feature_importance2 = pd.Series(importance2, index=X2.columns)
print(feature_importance2.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(y2_test, y2_rf_test_pred, alpha=0.3)
z = np.polyfit(y2_train, y2_rf_train_pred, 1)
p = np.poly1d(z)
plt.plot(y2_train, p(y2_train), "#F8766D")
plt.xlabel("Actual Renew_Share")
plt.ylabel("Predicted Renew_Share")
plt.show()
```
